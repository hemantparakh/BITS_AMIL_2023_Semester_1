```python
def generate_board():
    return [['-' for _ in range(6)] for _ in range(6)]

def check_game_over(grid):
    return all(all(piece != '-' for piece in row) for row in grid)

def detect_empty_positions(grid):
    return [(i, j) for i in range(len(grid)) for j in range(len(grid)) if grid[i][j] == '-']

def display_board(grid):
    for rows in grid:
        print(" ".join(rows))
    print()

def check_two_consecutive(grid, i, j, player):
    for move in ["east", "southeast", "south", "southwest", "west", "northwest", "north", "northeast"]:

        if move == "north" and i - 1 >= 0:
            two_consecutive = grid[i][j] == player and grid[i - 1][j] == player
            if two_consecutive:
                return two_consecutive
        elif move == "northeast" and i - 1 >= 0 and j + 1 < len(grid[i]):
            two_consecutive = grid[i][j] == player and grid[i - 1][j + 1] == player
            if two_consecutive:
                return two_consecutive
        elif move == "east" and j + 1 < len(grid[i]):
            two_consecutive = grid[i][j] == player and grid[i][j + 1] == player
            if two_consecutive:
                return two_consecutive
        elif move == "southeast" and i + 1 < len(grid) and j + 1 < len(grid[i]):
            two_consecutive = grid[i][j] == player and grid[i + 1][j + 1] == player
            if two_consecutive:
                return two_consecutive
        elif move == "south" and i + 1 < len(grid):
            two_consecutive = grid[i][j] == player and grid[i + 1][j] == player
            if two_consecutive:
                return two_consecutive
        elif move == "southwest" and i + 1 < len(grid) and j - 1 >= 0:
            two_consecutive = grid[i][j] == player and grid[i + 1][j - 1] == player
            if two_consecutive:
                return two_consecutive
        elif move == "west" and j - 1 >= 0:
            two_consecutive = grid[i][j] == player and grid[i][j - 1] == player
            if two_consecutive:
                return two_consecutive
        elif move == "northwest" and i - 1 >= 0 and j - 1 >= 0:
            two_consecutive = grid[i][j] == player and grid[i - 1][j - 1] == player
            if two_consecutive:
                return two_consecutive
    return False


def three_consecutive_count(grid, i, j, player):
    three_consecutive = 0
    for move in ["east", "southeast", "south", "southwest"]:
        if move == "east" and j + 2 < len(grid[i]):
            three_consecutive += grid[i][j] == player and grid[i][j + 1] == player and grid[i][j + 2] == player
        elif move == "southeast" and i + 2 < len(grid) and j + 2 < len(grid[i]):
            three_consecutive += grid[i][j] == player and grid[i + 1][j + 1] == player and grid[i + 2][j + 2] == player
        elif move == "south" and i + 2 < len(grid):
            three_consecutive += grid[i][j] == player and grid[i + 1][j] == player and grid[i + 2][j] == player
        elif move == "southwest" and i - 2 >= 0 and j + 2 < len(grid[i]):
            three_consecutive += grid[i][j] == player and grid[i - 1][j + 1] == player and grid[i - 2][j + 2] == player
    return three_consecutive


def opponent_3_consecutive(grid, i, j, player):
    for rival in ["X", "O", "T"]:
        if rival == player:
            continue

        ##3 consecutive coins on eastside
        if j + 3 < len(grid):
            if grid[i][j + 1] == rival and grid[i][j + 2] == rival and grid[i][j + 3] == rival:
                return True

        ##3 consecutive coins on southeast diagnol
        if i + 3 < len(grid) and j + 3 < len(grid):
            if grid[i + 1][j + 1] == rival and grid[i + 2][j + 2] == rival and grid[i + 3][j + 3] == rival:
                return True

        ##3 consecutive coins on southside
        if i + 3 < len(grid):
            if grid[i + 1][j] == rival and grid[i + 2][j] == rival and grid[i + 3][j] == rival:
                return True
        ##3 consecutive coins on southside left
        if i - 3 >= 0 and j - 3 >= 0:
            if grid[i - 1][j - 1] == rival and grid[i - 2][j - 2] == rival and grid[i - 3][j - 3] == rival:
                return True
        ##3 consecutive coins on left
        if j - 3 >= 0:
            if grid[i][j - 1] == rival and grid[i][j - 2] == rival and grid[i][j - 3] == rival:
                return True
        ##3 consecutive coins on northwest side
        if i - 3 >= 0 and j - 3 >= 0:
            if grid[i - 1][j - 1] == rival and grid[i - 2][j - 2] == rival and grid[i - 3][j - 3] == rival:
                return True
        ##3 consecutive coins on northside
        if i - 3 >= 0:
            if grid[i - 1][j] == rival and grid[i - 2][j] == rival and grid[i - 3][j] == rival:
                return True

        ##3 consecutive coins on northeast diagnol
        if i - 3 >= 0 and j + 3 < len(grid):
            if grid[i - 1][j + 1] == rival and grid[i - 2][j + 2] == rival and grid[i - 3][j + 3] == rival:
                return True

    return False


def opponent_2_consecutive(grid, i, j, player):
    for rival in ["X", "O", "T"]:
        if rival == player:
            continue

        ##2 consecutive coins on eastside
        if j + 2 < len(grid):
            if grid[i][j + 1] == rival and grid[i][j + 2] == rival:
                return True

        ##2 consecutive coins on southeast diagnol
        if i + 2 < len(grid) and j + 2 < len(grid):
            if grid[i + 1][j + 1] == rival and grid[i + 2][j + 2] == rival:
                return True

        ##2 consecutive coins on southside
        if i + 2 < len(grid):
            if grid[i + 1][j] == rival and grid[i + 2][j] == rival:
                return True
        ##2 consecutive coins on southside left
        if i - 2 > 0 and j - 2 > 0:
            if grid[i - 1][j - 1] == rival and grid[i - 2][j - 2] == rival:
                return True
        ##2 consecutive coins on westside
        if j - 2 > 0:
            if grid[i][j - 1] == rival and grid[i][j - 2] == rival:
                return True
        ##2 consecutive coins on northwest side
        if i - 2 > 0 and j - 2 > 0:
            if grid[i - 1][j - 1] == rival and grid[i - 2][j - 2] == rival:
                return True
        ##2 consecutive coins on northside
        if i - 2 > 0:
            if grid[i - 1][j] == rival and grid[i - 2][j] == rival:
                return True

        ##2 consecutive coins on northeast diagnol
        if i - 2 > 0 and j + 2 < len(grid):
            if grid[i - 1][j + 1] == rival and grid[i - 2][j + 2] == rival:
                return True

        return False


def opponent_3_consecutive_split(grid, i, j, player):
    return check_two_consecutive(grid, i, j, player) and opponent_2_consecutive(grid, i, j, player)

# Check if we have a winner of Tic Tac Toe
def detect_winner(grid, player):
    size = len(grid)

    # Any consecutive rows or column matches?
    for i in range(size):
        for j in range(size - 3):
            if all(grid[i][j + k] == player for k in range(4)):
                return True
            if all(grid[j + k][i] == player for k in range(4)):
                return True

    # Any consecutive diagonals matches?
    for i in range(size - 3):
        for j in range(size - 3):
            if all(grid[i + k][j + k] == player for k in range(4)):
                return True
            if all(grid[i + k][j + 3 - k] == player for k in range(4)):
                return True

    return False

def determine_next_virtual_player(virtual_player):
    if virtual_player == "X":
        next_player = "O"
    elif virtual_player == "O":
        next_player = "T"
    elif virtual_player == "T":
        next_player = "X"
    return next_player


def is_any_winner(board_state):
    return detect_winner(board_state, "X") or detect_winner(board_state, "O") or detect_winner(board_state, "T")

eval_tracker = 0

def determine_next_virtual_player_pos(virtual_player):
    if virtual_player == "X":
        return 0
    elif virtual_player == "O":
        return 1
    elif virtual_player == "T":
        return 2
```

### Implementation of the Min-Max algorithm

```python
def minimax(grid, depth, current_player, evaluating_player, alpha_beta):
    if depth == 0 or is_any_winner(grid) or check_game_over(grid):
        return evaluate_board(grid, evaluating_player)

    max_eval = (-float('inf'), -float('inf'), -float('inf'))
    for move in detect_empty_positions(grid):
        grid[move[0]][move[1]] = current_player
        next_player = determine_next_virtual_player(current_player)
        eval_value = minimax(grid, depth - 1, next_player, current_player, alpha_beta)
        grid[move[0]][move[1]] = '-'
        max_eval = tuple(max(max_eval[i], eval_value[i]) for i in range(3))

        alpha_beta = tuple(max(alpha_beta[i], eval_value[i]) for i in range(3))
        if alpha_beta[determine_next_virtual_player_pos(evaluating_player)] <= alpha_beta[determine_next_virtual_player_pos(current_player)] \
                and evaluating_player != current_player:
            break
        elif alpha_beta[determine_next_virtual_player_pos(evaluating_player)] > alpha_beta[
            determine_next_virtual_player_pos(current_player)] \
                and evaluating_player == current_player:
            break

    return max_eval
```

### Implementation of the alpha-beta pruning  

```python
def calculate_optimum_move(grid, virtual_player):
    best_val = (-float('inf'), -float('inf'), -float('inf'))
    best_move = (-1, -1)

    ai_player_index = determine_next_virtual_player_pos(virtual_player)
    alpha_beta = tuple(-float('inf') if i == ai_player_index else float('inf') for i in range(3))

    for move in detect_empty_positions(grid):
        grid[move[0]][move[1]] = virtual_player
        next_player = determine_next_virtual_player(virtual_player)
        move_val = minimax(grid, 2, next_player, virtual_player, alpha_beta)
        grid[move[0]][move[1]] = '-'

        if move_val[ai_player_index] > best_val[ai_player_index]:
            best_move = move
            best_val = move_val

        alpha_beta = tuple(
            max(alpha_beta[i], move_val[i]) if i == ai_player_index else min(alpha_beta[i], move_val[i]) for i in
            range(3))

    return best_move
```

### Choice and implementation of the Static Evaluation Function.

```python
def evaluate_board(grid, current_player):
    global eval_tracker
    eval_tracker += 1

    score_X = 0
    score_O = 0
    score_T = 0

    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j] == "-":
                continue

            for player in ['X', 'O', 'T']:
                if grid[i][j] != player:
                    continue

                # three_consecutive positions are strategically powerful, rank it accordingly by allocating a score of 100
                three_consecutive = three_consecutive_count(grid, i, j, player)
                if three_consecutive > 0:
                    if player == 'X':
                        score_X += 100 * three_consecutive
                    elif player == 'O':
                        score_O += 100 * three_consecutive
                    elif player == 'T':
                        score_T += 100 * three_consecutive

                # three_consecutive is also a stratigic position but lesser than three_consecutive so lets rank it by allocating a score of 50
                if check_two_consecutive(grid, i, j, player):
                    if player == 'X':
                        score_X += 50
                    elif player == 'O':
                        score_O += 50
                    elif player == 'T':
                        score_T += 50

                # three_consecutive positions of rival is strategic position we have to minimize the score of rival here, give 250 to block space next to winning position of rival
                if opponent_3_consecutive(grid, i, j, player) or opponent_3_consecutive_split(grid, i, j, player):
                    if player == 'X':
                        score_X += 250
                    elif player == 'O':
                        score_O += 250
                    elif player == 'T':
                        score_T += 250

                #two consecutive coins by rival, awards score of 20
                #helps to prioritize position that blocks oppenent from building a third coin.
                if opponent_2_consecutive(grid, i, j, player):
                    if player == 'X':
                        score_X += 20
                    elif player == 'O':
                        score_O += 20
                    elif player == 'T':
                        score_T += 20

                ##if computer player detects a winning move to itself then award highest score here, 1000 in this case
                if detect_winner(grid, player):
                    if player == 'X':
                        score_X += 1000
                    elif player == 'O':
                        score_O += 1000
                    elif player == 'T':
                        score_T += 1000

    return score_X, score_O, score_T
```

```python
def play_game():
    game_board = generate_board()

    print("Lets play Three-Player Tic-Tac-Toe (Players: X[Human], O[Computer1], T[Computer2])")
    display_board(game_board)

    while True:
        global eval_tracker
        eval_tracker = 0
        player_move = input("Enter move for player X in (row,col form 0-5): ")
        try:
            move = tuple(map(int, player_move.replace(',', ' ').split()))
        except ValueError:
            print("Invalid move! please enter valid row and column values.")
            continue

        if move not in detect_empty_positions(game_board):
            print("Invalid move! position not available for this move.")
            continue

        game_board[move[0]][move[1]] = 'X'

        print("Human Player[X] move:")
        display_board(game_board)

        if detect_winner(game_board, 'X'):
            print("********** Human Player[X] won the game **********")
            break
        elif check_game_over(game_board):
            print("It's a tiebreaker!")
            break

        # Computer move (Player O)
        move = calculate_optimum_move(game_board, 'O')
        game_board[move[0]][move[1]] = 'O'
        print("AI Player[O] move:")
        display_board(game_board)

        if detect_winner(game_board, 'O'):
            print("AI Player[O] won the game!")
            break
        elif check_game_over(game_board):
            print("It's a tiebreaker!")
            break

        # Computer move (Player T)
        move = calculate_optimum_move(game_board, 'T')
        game_board[move[0]][move[1]] = 'T'
        print("AI Player[T] move:")
        display_board(game_board)

        if detect_winner(game_board, 'T'):
            print("AI Player[T] won the game!")
            break
        elif check_game_over(game_board):
            print("It's a tiebreak!")
            break

if __name__ == "__main__":
    grid = generate_board()
    play_game()
```
